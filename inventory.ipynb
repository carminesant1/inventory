```python
import json

try:
    with open("inventario.json") as json_file:
        inventory = json.load(json_file)
except Exception as x:
    print("Primo accesso al software, i comandi disponibili sono i seguenti: \n -aggiungi: aggiungi un prodotto al magazzino \n -vendita: registra una vendita effettuata \n -profitti: mostra i profitti totali \n -elenca: elenca i prodotti in magazzino \n -aiuto: mostra i possibili comandi \n -chiudi: esci dal programma")
    inventory = {"Profitti" : [0.0, 0.0]}

def save():
  """
  Save the inventory in a JSON file "inventario.json"
  """
  with open("inventario.json", "w") as json_file:
    json.dump(inventory, json_file, indent=5)



error = "Il numero inserito non è valido"



def add_prod(prod=None, quant=None, cost=None, price=None):
  """
  Adds a product to inventory or updates its quantity if already present
  """
  prod=input("Nome del prodotto: ").lower()

  while quant is None or quant <= 0:
    try:
        quant = int(input("Quantità: "))
        if quant <= 0:
            raise ValueError(error)
    except ValueError:
        print(error)

  if prod not in inventory:

    while cost is None or cost <= 0:
      try:
        cost=float(input("Prezzo di acquisto: "))
        if cost <= 0:
            raise ValueError(error)
      except ValueError:
        print(error)

    while price is None or price <= 0:
      try:
        price=float(input("Prezzo di vendita: "))
        if price <= 0:
            raise ValueError(error)
      except ValueError:
        print(error)

  if prod in inventory :
    inventory[prod][0] += quant
  else:
    inventory[prod] = [quant, cost, price]

  inventory["Profitti"][1] += quant*inventory[prod][1]
  save()
  print(f"AGGIUNTO: {quant} x {prod}")



def sell_prod(prod=None, quant=None, list_prod=[], fin_price=[] ):
  """
  Print the sales receipt, when recording the transition consider whether the inventory is empty, whether it contains the product to be sold and whether the required quantity is present in the inventory
  """
  if len(inventory) > 1:
    prod = input("Nome del prodotto: ").lower()
    while prod not in inventory:
      print("Prodotto non disponibile!")
      prod = input("Nome del prodotto: ").lower()

    while quant is None or quant > inventory[prod][0] or quant <=0:
      try:
          quant = int(input("Quantità: "))
          if quant > inventory[prod][0] or quant <=0:
              raise ValueError
      except ValueError:
          print(f"{error}\nN.B. : Ci sono {inventory[prod][0]} unità di {prod}!")

    inventory[prod][0] -= quant
    inventory["Profitti"][0] += quant*inventory[prod][2]
    list_prod.append(f"{quant} x {prod} : €{inventory[prod][2]}")
    fin_price.append(quant*inventory[prod][2])
    if inventory[prod][0] == 0:
      del inventory[prod]
    oth_prod = None
    while oth_prod is None:
      try:
        oth_prod = input("Aggiungere un altro prodotto? (sì/no) ").lower()
        if oth_prod not in ("si", "sì", "no"):
          raise Exception
      except Exception:
        print("Inserire sì/no")
        oth_prod = None
    if oth_prod in ("si,sì"):
      sell_prod()
    else:
      save()
      print("VENDITA REGISTRATA")
      for i in range(len(list_prod)):
        print(list_prod[i])
      print(f"Totale: €{sum(fin_price): .2f}")
      list_prod.clear()
      fin_price.clear()

  else:
    print("L'inventario è vuoto!")
    if len(list_prod) > 0:
      save()
      print("VENDITA REGISTRATA")
      for i in range(len(list_prod)):
        print(list_prod[i])
      print(f"Totale: €{sum(fin_price): .2f}")
      list_prod.clear()
      fin_price.clear()



def list_prod():
  """
  Shows the products and their quantities in inventory
  """
  inventory_prod = inventory.copy()
  del inventory_prod["Profitti"]
  print("Prodotto  Quantità  Prezzo")
  for i in inventory_prod:
    print(f"{i}   {inventory_prod[i][0]}   €{inventory_prod[i][2]}")



def profit():
  """
  Shows gross profit (revenue) and net profit (revenue - costs)
  """
  net_profit = inventory["Profitti"][0]
  gross_profit = inventory["Profitti"][0] - inventory["Profitti"][1]
  print(f"Profitto lordo: €{net_profit: .2f} \nProfitto netto: €{gross_profit: .2f}")



def help():
  """
  Shows the commands available to the user
  """
  print("I comandi disponibili sono i seguenti: \n -aggiungi: aggiungi un prodotto al magazzino \n -vendita: registra una vendita effettuata \n -profitti: mostra i profitti totali \n -elenca: elenca i prodotti in magazzino \n -aiuto: mostra i possibili comandi \n -chiudi: esci dal programma")



while True:
  com=input("Inserisci un comando: ").lower()

  if com=="aggiungi":
    add_prod()

  elif com == "vendita":
    sell_prod()

  elif com=="elenca":
    list_prod()

  elif com=="profitti":
    profit()

  elif com=="chiudi":
    save()
    print("Bye bye")
    break

  elif com == "aiuto":
    help()

  else:
    print("Comando non valido!")
    help()
```
